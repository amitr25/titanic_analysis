```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
x.head()
```

```python
x.tail()
```

```python
Q1 = x['duration'].quantile(0.25)
Q3 = x['duration'].quantile(0.75)
IQR = Q3 - Q1
```

```python
IQR
```

```python
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
```

```python
x_no_outliers = x[(x['duration'] >= lower_bound) & (x['duration'] <= upper_bound)]
```

```python
x_no_outliers.shape
```

```python
plt.figure(figsize=(12, 5))
```

```python
plt.subplot(1, 2, 1)
sns.boxplot(x=x_no_outliers['duration'])
plt.title("Boxplot of Duration (Outliers Removed)")
```

```python
plt.subplot(1, 2, 2)
sns.histplot(x_no_outliers['duration'], kde=True)
plt.title("Distribution Plot of Duration (Outliers Removed)")
```
